Q4. Explain the use of the groupby() method in Pandas.

Answer:


The groupby() method in Pandas is a powerful tool used for splitting data into groups based on some criteria, performing operations on each group independently, and then combining the results back into a single structure. Itâ€™s commonly used for aggregation, transformation, and filtering tasks.

```python
#Example
import pandas as pd

# Sample DataFrame
data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Sales': [100, 200, 150, 250, 130, 300]
}

df = pd.DataFrame(data)

# Group by 'Store' and calculate the total sales for each store
grouped = df.groupby('Store')['Sales'].sum()
print(grouped)
```

Q15. How does NumPy handle multidimensional arrays?

Answer:

NumPy handles multidimensional arrays (ndarrays) by providing efficient tools for creating, manipulating, and performing operations on arrays with multiple dimensions. Key features include:

1.   Array Creation: Multidimensional arrays can be created using nested lists and the np.array() function.
2.   Shape and Dimensions: The shape of arrays is easily accessible with .shape, and the number of dimensions can be checked using .ndim.
3. Indexing and Slicing: NumPy allows advanced indexing and slicing for extracting subarrays or individual elements across multiple dimensions
4. Broadcasting: NumPy supports broadcasting, enabling operations on arrays of different shapes by automatically adjusting their dimensions.
5. Vectorized Operations: Arrays support fast, element-wise operations without the need for explicit loops.
6. Reshaping: Arrays can be reshaped into different dimensions using .reshape(), while keeping the data intact.
7. Flattening: Use .flatten() or .ravel() to convert arrays to 1D.
8. Efficient Storage: NumPy arrays store data contiguously in memory, offering high performance and efficient memory usage.
9. Advanced Manipulation: It provides functions for transposing, concatenating, stacking, and splitting arrays.

Q18. What are some advanced features of NumPy?

Answer:

**Broadcasting**: Allows operations between arrays of different shapes by automatically expanding the smaller array to match the larger one, enabling efficient element-wise operations without explicit loops.

**Vectorization**: Enables fast element-wise operations without using Python loops by performing computations at the C-level, leading to significant performance improvements.

**Advanced Indexing**: Supports fancy indexing, boolean indexing, and using arrays to index other arrays, providing flexible ways to access and manipulate data.

**Memory Views**: Enables efficient manipulation of data by creating views (not copies) of arrays, allowing modifications without duplicating memory.

**Structured Arrays**: Allows arrays with heterogeneous data types (like a mix of integers, floats, and strings), similar to records or tuples, and is useful for storing complex data.

**Random Module**: Provides advanced random number generation capabilities, supporting various distributions like normal, uniform, and binomial, and useful for simulations and statistical analysis.

**Linear Algebra**: Offers comprehensive linear algebra functions such as matrix multiplication, eigenvalue decomposition, and solving linear systems, essential for scientific computing and machine learning.

**Vectorized String Operations**: Enables efficient element-wise string manipulation using the np.char module, useful for operations like string replacement, formatting, and transformation.

Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

Answer:

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load dataset
iris = sns.load_dataset('iris')

# Step 3: Calculate correlation matrix
correlation_matrix = iris.corr()

# Step 4: Create a heatmap using Seaborn
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Step 5: Display the plot
plt.show()
```

Q8. Create a line plot with multiple lines using Matplotlib.

Answer:



```python
import matplotlib.pyplot as plt

# Data for the lines
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 4, 6, 8, 10]
y3 = [5, 7, 9, 11, 13]

# Create the plot
plt.figure(figsize=(8, 6))

# Add lines to the plot
plt.plot(x, y1, label='y = x^2', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='y = 2x', color='red', linestyle='--', marker='s')
plt.plot(x, y3, label='y = 2x + 3', color='green', linestyle='-.', marker='^')

# Add title and labels
plt.title('Multiple Line Plot Example', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)

# Add a legend
plt.legend(fontsize=12)

# Display the grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```

Q9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

Q10. Create a histogram using Seaborn to visualize a distribution.

Answer:






```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.normal(loc=50, scale=10, size=500)  # Normal distribution (mean=50, std=10)

# Create a histogram with Seaborn
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.7)

# Add labels and a title
plt.title('Histogram of Data Distribution', fontsize=16)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

Q11. Perform matrix multiplication using NumPy.

Answer:




```python
import numpy as np

# Define two matrices
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Perform matrix multiplication
result = np.dot(A, B)  # or equivalently, A @ B in Python 3.5+

# Display the matrices and result
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nResult of A x B:")
print(result)
```

Q13. Create a 3D scatter plot using Plotly.

Answer:

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate some sample data
np.random.seed(42)
data = {
    'X': np.random.rand(50) * 10,
    'Y': np.random.rand(50) * 10,
    'Z': np.random.rand(50) * 10,
    'Category': np.random.choice(['A', 'B', 'C'], size=50)  # Optional categorical data
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a 3D scatter plot
fig = px.scatter_3d(
    df,
    x='X',
    y='Y',
    z='Z',
    color='Category',  # Color points by category
    size_max=10,
    title='3D Scatter Plot Example'
)

# Show the plot
fig.show()
```
